```python
# Підгружаю необхідні бібліотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Підготовка даних

```python
# Зчитую файл
url = "https://docs.google.com/spreadsheets/d/1mAP0_Noi8f9-Ha4et5TlksabocHsPtUM7EalYH5xX_o/export?format=csv&gid=0"

# ПРибираю назви стовбців, шоб побачити де закінчується перша таблиця
data = pd.read_csv(url, header=None)
data
```

```python
# Видаляю 4 стовбикю, він пустий
data = data.drop(columns=[4])
data
```

```python
# Копіюю датафрейм, змінюю назви стовбчиків як має бути, за допомогою слайсу видаляю рядок з назвами колонок
df = data.copy()
df.columns = df.iloc[0]
df = df[1:]
df
```

```python
# ВИдалимо порожні рядки і коментар
df.loc[df['Subs type'] == 'Subscription is renewed automatically\t\t\t', 'Subs type'] = np.nan
df = df.dropna(how="all")
df
```

```python
# Перевірю тип данних в таблиці
df.info()
```

```python
# ПРиводжу стовбчики до правильних типів даних
df[['user_id', 'subscription_renewal_amount']] = df[['user_id', 'subscription_renewal_amount']].astype(int)
df['first_event_date'] = pd.to_datetime(df['first_event_date'])
df['Trial period_days'] = df['Trial period'].str.extract('(\d+)').astype(float)
df['Subscription period_month'] = df['Subscription period '].str.extract('(\d+)').astype(float)
df.drop(columns=['Trial period', 'Subscription period '], inplace=True)
df['First payment_doll'] = df['First payment'].str.replace('$', '', regex=False).astype(float)
df['Rebill payment__doll'] = df['Rebill payment'].str.replace('$', '', regex=False).astype(float)
df.drop(columns=['First payment', 'Rebill payment'], inplace=True)
df.info()
```

```python
df
```

```python
# На початку не думав розюивати на 2 таблиці, але всеж-таки розібю
# Перша таблиця — усі рядки, але тільки перші 4 колонки
df_users = df.iloc[:, :4].copy()
df_users
```

```python
# Друга таблиця — тільки перші 2 рядки, починаючи з 4-ї колонки
df_types = df.iloc[:2, 4:].copy()
df_types
```

## Завдання_1

Тут, думаю, треба поснити як я думаю. Ті хто взяв пробний період заплатили 6,99 за 7 днів користування сервісом, наступна оплата помісячно з вартістю 29,99

$$
\text{Total revenue}_{trial} = \sum_{i=1}^{N}
\Big( \text{First payment} + \text{Rebill payment} \times \text{subscription_renewal_amount}_i \Big)
$$


```python
first_payment = df_types.iloc[0, 3]
rebill_payment = df_types.iloc[0, 4]
```

```python
# Порахую кількість людей хто взяв пробну підписку
df_trial = df_users[df_users['is_trial'] == '0'].copy()
```

```python
# Чисто випадково вирішив перевірити на дублікати
df_trial['user_id'].value_counts().head(10)
```

```python
# Це вже цікаво, цнікальних корстуваці то тільки 980, тобто з кожним платежем рядок до таблиці додавався
df_trial['user_id'].nunique()
```

```python
df_trial[df_trial['user_id'] == 23712]
```

```python
# Групуємо дані за користувачем і беремо максимум поновлень
df_trial_unique = df_trial.groupby('user_id', as_index=False)['subscription_renewal_amount'].max()
df_trial_unique
```

```python
# Рахую індивідуальний дохід
df_trial_unique['user_revenue'] = first_payment + rebill_payment * df_trial_unique['subscription_renewal_amount']
```

```python
df_trial_unique
```

```python
# Загальний результат
total_trial_revenue = df_trial_unique['user_revenue'].sum()
n_trial_users = len(df_trial_unique)

print(f"Кількість унікальних trial-користувачів: {n_trial_users}")
print(f"Сукупний дохід від trial-користувачів: ${total_trial_revenue:,.2f}")
```

*Висновок*:

Було проведено очищення даних від дублікатів та коректно розраховано індивідуальний і загальний дохід від користувачів, які оформили підписку з пробним періодом.

**Кількість унікальних trial-користувачів: 980**

**Сукупний дохід від trial-користувачів: $193,627.92**

## Завдання 2

### Частина "а"

```python
df_trial['subscription_renewal_amount'].nunique()
```

```python
# Рахуємо кількість користувачів для кожного числа поновлень
renewal_counts = (df_trial_unique['subscription_renewal_amount'].value_counts().sort_index().reset_index())

renewal_counts.columns = ['Number of renewals', 'Number of users']
renewal_counts
```

```python
fig = px.bar(
    renewal_counts,
    x='Number of renewals',
    y='Number of users',
    title='Кількість користувачів за кількістю ребілів (Trial-підписки)',
    color='Number of renewals',
    text='Number of users',
    color_continuous_scale='Viridis'
)

fig.update_traces(texttemplate='%{text}', textposition='outside')
fig.update_layout(
    xaxis_title='Кількість ребілів',
    yaxis_title='Кількість користувачів',
    showlegend=False,
    height=500
)
fig.show()
```

### Частина "б"

```python
revenue_by_segment = (
    df_trial_unique
    .groupby('subscription_renewal_amount')['user_revenue']
    .sum()
    .reset_index()
)

fig = px.bar(
    revenue_by_segment,
    x='subscription_renewal_amount',
    y='user_revenue',
    color='user_revenue',
    text='user_revenue',
    color_continuous_scale='Plasma',
    title='Сукупний дохід за кількістю ребілів (Trial-підписки)'
)

fig.update_traces(texttemplate='$%{text:.2f}', textposition='outside')
fig.update_layout(
    xaxis_title='Кількість ребілів',
    yaxis_title='Сумарний дохід ($)',
    showlegend=False,
    height=500
)
fig.show()
```

```python
# Перевірив які роки включені в даний датасет
df_trial['year'] = df_trial['first_event_date'].dt.year
df_trial['year'].unique()
```

```python
# Перевірка пероду данних
print("Найраніша дата:", df_trial['first_event_date'].min())
print("Найпізніша дата:", df_trial['first_event_date'].max())
```

```python
# беремо лише останній запис для кожного користувача
df_trial_last = df_trial.sort_values('subscription_renewal_amount').groupby('user_id').tail(1)

# рахуємо правильний дохід
df_trial_last['user_revenue'] = first_payment + rebill_payment * df_trial_last['subscription_renewal_amount']

# групуємо за датою
revenue_timeline = (
    df_trial_last
    .groupby('first_event_date')['user_revenue']
    .sum()
    .reset_index()
    .sort_values('first_event_date')
)
```

```python
fig = px.line(
    revenue_timeline,
    x='first_event_date',
    y='user_revenue',
    text='user_revenue',
    color_discrete_sequence=['#1f77b4'],
    title='Динаміка доходу у часі (Trial-підписки)',
    markers=True
)

fig.update_traces(
    textposition='top center',   # 👈 підпис над точкою
    texttemplate='%{text:,.0f}$' # 👈 формат — без десяткових, з $
)

fig.update_layout(
    xaxis_title='Дата',
    yaxis_title='Сумарний дохід ($)',
    height=500,
)
fig.show()
```
